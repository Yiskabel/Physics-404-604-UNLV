```python
#Problem no.1 

#This is the graph of y = f_n/(f_n - 1) as a function of n in the range before
#python refuses to calculate f_n. The plot is further truncated to show how 
#the function approaches the limit of F = 1.


import pylab as p
from numpy import *

def fib3(n, table): #This function defines the fibonacci sequence and was worked through in class
    if n==0 or n==1:
        table[n] = n
    if table[n] is None:
        table[n] = fib3(n-1, table) + fib3(n-2, table)
    return table[n]

x = [0] #I've initialized the list of x and y at the trivial value of (0,0)
y = [0] #because of range restrictions in the values for 1 and 2.
for n in range(3, 2966, 1):                 #the range starts at 3 because the values of 1 and 2 yield infinity in the y function
    table = [None]*(n+1)
    f = fib3(n, table)/(fib3(n, table) - 1)
    x.append(n)
    y.append(f)

#print(x,y)
        
p.title('Plot of $f_n/(f_n - 1)$ vs. n')
p.xlabel('n')
p.ylabel('$f_n/(f_n - 1)$')    
p.scatter(x,y)
p.xlim([0,30]) #Limits the x-axis of the plot; could've just made for loop shorter.
p.grid(True)
p.show()
```

```python
#Problem no2

#Here is the program that generates (n,time) for iteration method and saves the pairs into a text file

def fib2(n):
    if n == 0: return 0
    elif n == 1: return 1
    else: return fib2(n-1) + fib2(n-2)
    
import time
n = 1
file1 = open("n_time_iteration", "w+")
for n in range(1,31,1):
    start_time = time.time()
    x = fib2(n)
    y = time.time() - start_time
    #print(x,y)
    file1.write("%s " % n)
    file1.write("%s\n" % y)
file1.close()
print("saved as n_time_iteration.txt")
```

```python
#This program extracts the data from the iteration text file, and plots log(time) vs log(n) after convertings the strings to the
#correct variable type.

import pylab as p
from numpy import *

read1 = open("n_time_iteration" , "r")

x = []
y = []
for line in read1:
    n, time = line.split(" ")
    x1 = log10(int(n))
    y1 = log10(float(time))
    x.append(x1)
    y.append(y1)
    
p.title('Iteration: log(time) vs log(n)')
p.xlabel('log(n)')
p.ylabel('log(Time)')  
p.plot(x,y)
p.show()
```

```python
#Here is the program that generates (n,time) for tabulation method and saves the pairs into a text file


def fib4(n):
    table=[0]*(n+1)
    table[1] = 1
    
    for i in range(2,n+1):
        table[i] = table[i - 1] + table[i - 2]
        
    return table[n]

import time
n=1
file2 = open("n_time_tabulation", "w+")
for n in range(1,100000,1000):
    start_time = time.time()
    x = fib4(n)
    y = time.time()-start_time
    #print(x,y)
    file2.write("%s " % n)
    file2.write("%s\n" % y)
file2.close()
print("saved as n_time_tabulation.txt")
```

```python
#Here is the program that generates (n,time) for matrix method and saves the pairs into a text file, the function for the 
#matrix fibonacci sequence was taken from lecture notes.

def arrpow(arr, n):
    yarr=arr
    if n<1:
        print('n needs to be larger than 1')
    if n==1:
        return arr
    yarr = arrpow(arr, n//2)
    yarr = [[yarr[0][0]*yarr[0][0]+yarr[0][1]*yarr[1][0],yarr[0][0]*yarr[0][1]+yarr[0][1]*yarr[1][1]],
            [yarr[1][0]*yarr[0][0]+yarr[1][1]*yarr[1][0],yarr[1][0]*yarr[0][1]+yarr[1][1]*yarr[1][1]]]
    if n%2:
        yarr=[[yarr[0][0]*arr[0][0]+yarr[0][1]*arr[1][0],yarr[0][0]*arr[0][1]+yarr[0][1]*arr[1][1]],
            [yarr[1][0]*arr[0][0]+yarr[1][1]*arr[1][0],yarr[1][0]*arr[0][1]+yarr[1][1]*arr[1][1]]]
    return yarr

def fib5(n):
    arr= [[1,1],[1,0]]
    f=arrpow(arr,n-1)[0][0]
    return f

import time
n=2
file3 = open("n_time_matrix", "w+")
for n in range(2,1000000,10000):
    start_time = time.time()
    x = fib5(n)
    y = time.time()-start_time
    #print(x,y)
    file3.write("%s " % n)
    file3.write("%s\n" % y)
file3.close()
print("saved as n_time_matrix.txt")
```
