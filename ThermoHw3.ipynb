```python
# Question 1) Here i can calculate the multiplicity of outcomes given a set of N coins. I've defined the factorial function
#explicitly and I've created a function that inputs the total number of Coins and Returs 2 sets. The first set is the number
#of heads appearing and the second set is the multiplicity of that configuration.

import pylab as p
from numpy import *

def factorial(x):
    if x == 0:
        return 1
    else:
        if x < 2000: 
            return x*factorial(x-1)
        else:
            return (2*pi*x)**(0.5)*(x/e)**x
    
def mult(N):
    x = []
    y = []
    for i in range(0,N+1):
        x.append(i)
        y.append(factorial(N)/(factorial(i)*factorial(N-i)))
    return x,y    
```

```python
N = 10
[x,y] = mult(N)
p.bar(x,y)
p.xlabel("number of heads flipped")
p.ylabel("multiplicity")
p.title("N = 10")
p.show()
```

```python
#Question 2) Here i calculate and plot the multiplicty of energy levels for an einstein solid of given size N. I've created
# a new multiplicty function that only works when looking at energy levels starting from 0 and going to 10. The function 
# returns the individual energy level and then the multiplicty of that energy level.
def mult_E(N):
    x = []
    y = []
    for q in range(0,11):
        x.append(q)
        y.append(factorial(N+q-1)/(factorial(q)*factorial(N-1)))
    return x,y
```

```python
N = 10
[x,y] = mult_E(N)
p.bar(x,y)
p.xlabel("number of energy levels q")
p.ylabel("multiplicity")
p.title("N = %g" % N)
p.show()
```

```python
# Problem 3) In the following, i had to use the library function for factorials because my code was not efficient enough for large N. 
#The results are as expected centered about 2/3 of q because the ratio of N2 to N1 is 2/3. As usual the code outputs
#one array counting the values of q being used in N1, and spits out an array with the product of the multiplicities of 
#N1 and N2 for any given arrangement of q.

def mult_E2(N1,N2):
    x = []
    y = []
    for q in range(0,101):
        x.append(q)
        a = math.factorial(N1+q-1)/(math.factorial(q)*math.factorial(N1-1))
        b = math.factorial(N2+100-q-1)/(math.factorial(100-q)*math.factorial(N2-1))
        y.append(a*b)
    return x,y

N1 = 300
N2 = 200
[x,y] = mult_E2(N1,N2)
p.bar(x,y)
p.xlabel("number of energy levels q")
p.ylabel("multiplicity")
p.title("N1 = %g and N2 = %g" % (N1,N2))
p.show()
```
